```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
v = np.loadtxt("output.txt")
```

```python
maxwell = stats.maxwell
uniform = stats.uniform
```

# 2.1) 

Distribuci칩n uniforme (20 colisiones y temperatura m치s baja)

```python
_, bins, _ = plt.hist(v, 20, density=1, alpha=0.5)
param1,param2 = uniform.fit(v, floc=0)
best_fit_line = uniform.pdf(bins, param1, param2)
plt.plot(bins, best_fit_line)
```
